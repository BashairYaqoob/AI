```python
n=int(input("Enter the number of n"))
list1=[]
for i in range(n):
    inputn = int(input(f"Enter number {i+1}: "))
    list1.append(inputn)

print("\nOriginal list: ", list1)
counteven=0
countodd=0
for inputn in list1:
    if inputn%2==0:
        counteven+=1
    else:
        countodd+=1

print("Number of even numbers: ", counteven)
print("Number of odd numbers: ", countodd)

index=int(input("Enter index to replace (0-n): "))
if 0<= index <n:
    newval= int(input("Enter new value: "))
    list1[index]=newval
    print("Updated list: ", list1)
else:
    print("Invalid Index!")
```

```python
def salary(a):
    hra = 0.1*a
    da = 0.05*a
    return (a+hra+da)

a=float(input("Enter basic salary: "))
print("Total salary: $", salary(a))
```

```python
def stringzz(b):
    dict2={
        'Total characters' : len(b),
        'Uppercase letters' : 0,
        'Lowercase letters' : 0,
        'Number of Digits' : 0
    }

    for char in b:
        if char.isupper():
            dict2['Uppercase letters']+=1
        elif char.islower():
            dict2['Lowercase letters']+=1
        elif char.isdigit():
            dict2['Number of Digits']+=1
    return dict2

strinput = input("Enter a string: ")

for key, value in stringzz(strinput).items():
    print(f"{key} : {value}")
    
```

```python
userinput = input("Enter a string: ")
maxchar=''
maxlength=0
currchar=''
currlength=0

for char in userinput:
    if char==currchar:
        currlength+=1
    else:
        currchar=char
        currlength=1

    if currlength>maxlength:
        maxlength=currlength
        maxchar=currchar

longestss=maxchar*maxlength
print(f"Longest substring of consecutive identical characters: '{longestss}'")
print(f"Its length: {maxlength}")
```

```python
def repeats(c):
    repeated=[]
    checked=[]

    for num in c:
        if num not in checked:
            count=0
            for n in c:
                if n==num:
                    count+=1
            if count>1:
                repeated.append(num)
            checked.append(num)
    return repeated

numz=int(input("Enter number of n"))
list2=[]
for i in range(numz):
    inn= int(input(f"Enter number {i+1}: "))
    list2.append(inn)
print("\nOriginal list:", list2)
repeated_numbers = repeats(list2)
if repeated_numbers:
    print("Repeated numbers:", repeated_numbers)
else:
    print("No numbers are repeated.")
```
